```python
import os
from collections import Counter
import numpy as np

import matplotlib.pyplot as plt
from tqdm import tqdm


path = "/raid/home/rajivratn/hemant_rajivratn/librispeech/stable_checkpoint_181_250000/fresh/iter3label/label/multiple_layers_960/"
files = "train.km1000 train.km125 train.km25 train.km250 train.km50 train.km500".split(" ")
# files = "train.km25".split(" ")
```

```python
for f in files:
    print(f)
    
    with open(path + f, "r") as file:
        lines = file.readlines()
    lines = [str(i) for line in tqdm(lines) for i in line.strip().split()]
    # join the lines with a space
    
    
    print(f"len(lines): {len(lines)}")
    char_counts = Counter(lines)  # Example output: [('a', 2), ('d', 1)]
    char_counts = dict(char_counts)
    vocab = list(char_counts.keys())
    
    print(f"char_counts: {char_counts}")
    
    c = list(char_counts.values())
    # sort 
    c = sorted(c, reverse=True)
    c = np.array(c, dtype=np.float32)
    c /= c.sum()  # Normalize the counts to get probabilities
    
    cdf = np.cumsum(c)
    # keep only the top k until the cumulative sum is less than 0.95
    k = 0
    while k < len(cdf) and cdf[k] < 0.95:
        k += 1
        if k > 30:
            break
    cdf = cdf[:k]
    vocab = vocab[:k]
    c = c[:k]
    
    
    # Plotting the histogram
    plt.figure(figsize=(16, 6))
    plt.plot(cdf)
    plt.bar(vocab, c, color='blue', alpha=0.7, )
    plt.xlabel('Codebook Entry (Char)')
    plt.ylabel('Probability')
    plt.title('Codebook Usage Distribution')
    plt.grid(axis='y')
    plt.xticks([])
    plt.savefig(f"codebook_usage_{f}.png", dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()
    
```

```python
import random
def add_question_marks(texts=[]):
    
    modified_texts = []
    for sentence in tqdm(texts):
        modified_sentence = ['?']# Add question marks at start 
        previous_char = None
        for char in sentence:
            # if  char == previous_char insert a question mark
            if previous_char == char:
                modified_sentence.append("?")
            
            modified_sentence.append(char)

            # Randomly insert question marks with 0.25 probability
            if random.random() < 0.25 and modified_sentence[-1] != '?':
                modified_sentence.append("?")
            
            previous_char = char
                
        if modified_sentence[-1] != '?': 
            modified_sentence.append("?")  # Add a question mark at the end
        modified_texts.append("".join(modified_sentence))
    
    return  modified_texts
```

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
from collections import Counter


for sz in [25, 50, 125, 250, 500, 1000]:
    print(f"sz: {sz}")

    # Step 1: Initialize tokenizer
    tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
    tokenizer.pre_tokenizer = Whitespace()


    with open("/raid/home/rajivratn/hemant_rajivratn/librispeech/data/manifest/train.wrd", "r") as file:
        lines = file.readlines()
    lines = [line.strip() for line in lines]
    lines = add_question_marks(lines)

    # Step 2: Train tokenizer
    trainer = BpeTrainer(vocab_size=sz, special_tokens=["[UNK]"])
    tokenizer.train_from_iterator(lines, trainer=trainer)

    # Step 3: Tokenize all sentences and collect token IDs
    token_ids = []
    for sent in tqdm(lines):
        encoding = tokenizer.encode(sent)
        token_ids.extend(encoding.ids)

    # Step 4: Count token ID occurrences
    token_counts = Counter(token_ids)
    total = sum(token_counts.values())
    c = {token_id: count / total for token_id, count in token_counts.items()}
    c = list(c.values())
    c = sorted(c, reverse=True)
    c = np.array(c, dtype=np.float32)
    cdf = np.cumsum(c)
    
    # keep only the top k until the cumulative sum is less than 0.95
    k = 0
    while k < len(cdf) and cdf[k] < 0.95:
        k += 1
        if k > 30:
            break
    cdf = cdf[:k]
    c = c[:k]

    # Plotting the histogram
    plt.figure(figsize=(16, 6))
    plt.plot(cdf)
    plt.bar(range(k), c, color='blue', alpha=0.7, )
    plt.xlabel('Codebook Entry (Char)')
    plt.ylabel('Probability')
    plt.title('Codebook Usage Distribution')
    plt.grid(axis='y')
    plt.xticks([])
    plt.savefig(f"codebook_usage_chars_{sz}.png", dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()
```
